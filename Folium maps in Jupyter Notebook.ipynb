```python
import folium
from IPython.display import HTML
from IPython.display import IFrame

def display(map, height=300):
    '''Parameters: a folium instance - embeds map in HTML '''
    #m._build_map()
    m = map.get_root().render()# https://github.com/python-visualization/folium/issues/781#issuecomment-347660835
    
    #print(m)
    #srcdoc = m.HTML.replace('"','&quot;') #encoding of quotes
    srcdoc = m.replace('"','&quot;') #encoding of quotes    
    embed = HTML('<iframe srcdoc="{0}">'
                'style="width: 100%; height: {1}px; '
                'border: none"></iframe>'.format(srcdoc, height))
#     embed = IFrame('<iframe srcdoc="{0}">'
#                 'style="width: 100%; height: {1}px; '
#                 'border: none"></iframe>'.format(srcdoc, height),300,300)

    return embed
```

```python
# using same variable name 'map' as in function - maybe unwise
map = folium.Map(location=[51,-4], zoom_start=7, tiles='Mapbox Bright')

# adding markers
map.add_child(folium.Marker([50.9,-3.9], popup="Hello, I'm a marker", icon=folium.Icon(color='green')))

# use feature group in adding markers
fg = folium.FeatureGroup(name="PJ Map")
fg.add_child(folium.Marker([50.8,-3.8], popup="Hello, I'm a marker in a feature group #1", icon=folium.Icon(color='green')))
fg.add_child(folium.Marker([50.7,-4.1], popup="Hello, I'm a marker in a feature group #2", icon=folium.Icon(color='green')))
map.add_child(fg)

display(map)
```
